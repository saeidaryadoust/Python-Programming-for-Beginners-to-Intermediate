```python
Atr = []
Price = []
Link = []

for j in range(10):
    page = j+1
    site = requests.get(f"https://www.digikala.com/search/category-perfume/?q=%D9%85%D8%B1%D8%AF%D8%A7%D9%86%D9%87&pageno={page}&sortby=22")
    soup = BeautifulSoup(site.text, 'html.parser')
    
    x = soup.find('div' , {'class': 'o-page__content'})
    x2 = x.find_all('div', {'class': 'c-product-box__content--row'})
    
    for i in range(len(x2)):

        x3 = x2[i].find('a').text
        Atr.append(x3)

        x4 = x.find_all('div', {'class': 'c-price'})
        x4 = x4[i].find('div', {'class': 'c-price__value-wrapper'}).text
        x4 = x4.replace('\n', '')
        x4 = x4.replace(' ', '')
        x4 = x4.replace('تومان', '')
        x4 = x4.replace(',', '')
        Price.append(int(x4))
        
        x5 = x2[i].find('a')['href']
        x5 = "https://www.digikala.com" + x5
        Link.append(x5)
```

```python
print(f"Atr: {len(Atr)}")
print(f"Price: {len(Price)}")
print(f"Link: {len(Link)}")
```

```python
Dict = {
    'Atr': Atr,
    'Price': Price,
    'Link': Link
}
```

```python
##########################################################################################
```

```python
Book = []
Link = []
Date = []

url_site = "https://www.parsbook.com/page/"

for i in range(10):
    page = i+1
    site = requests.get(url_site + str(page))
    soup = BeautifulSoup(site.text, 'html.parser')
    
    x = soup.find_all('div' , {'class': 'singlebox purple'})
    
    for j in range(len(x)):
        book = x[j].find('a').text
        Book.append(book)

        link = x[j].find('a')['href']
        Link.append(link)

        x2 = soup.find_all('div' , {'class': 'singlebox date'})
        x2 = x2[j].text
        x2 = x2.replace('\n', '')
        x2 = x2.replace('تاریخ :', '')
        Date.append(x2)
```

```python
len(Date)
```

```python
Dict = {
    'Book': Book,
    'Date': Date,
    'Link': Link
}
```

```python
df = pd.DataFrame(Dict)
df.to_excel("E:\\dataset\\parsbook.xlsx")
```

```python
url_site = "https://taaghche.com/book-review/28544/%D8%A7%D8%B3%D8%AA%D8%A7%D8%B1%D8%AA%D8%A7%D9%BE-%D9%86%D8%A7%D8%A8?p="
Comments = []

for i in range(2):
    page = i+1
    site = requests.get(url_site + str(page))
    soup = BeautifulSoup(site.text, 'html.parser')
    x = soup.find('div' , {'class': 'bookComments_comments__vupiE container'})
    x2 = x.find_all('div' , {'class': 'bookComments_comment__32DTt'})

    for j in range(len(x2)):
        x3 = x2[j].text
        x3 = x3.replace('\n', '.')
        x3 = x3.replace('\u200c', ' ')
        Comments.append(x3)
```

```python
len(Comments)
```

```python
df = pd.DataFrame(Comments)
df.to_excel("E:\\dataset\\taaghche.xlsx")
```
